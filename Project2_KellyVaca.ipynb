# **Amazon's Top 100 Books**
## 1. Top 5 Most Popular Genres


```sql
SELECT
    book_genre AS Genre,
    COUNT(*) AS "Frequency"
FROM
    dbo.Book_Ranks
GROUP BY
    book_genre
HAVING
    COUNT(*) > 8
ORDER BY
    "Frequency" DESC;
```

**Explanation:**Â 

_This query finds the top 5 genres based on frequency within the ranked list. From the data, it is clear that Children's Books dominates the list, followed by Holiday Books._

## 2\. Average Rating By Genre

```sql
SELECT
    book_genre AS Genre,
    ROUND(AVG(book_rating),2) AS "Average Rating"
FROM
    dbo.Book_Ranks
GROUP BY
    book_genre
ORDER BY
    "Average Rating" DESC;
```

**Explanation:**

_This query explores the average ratings of books in each genre. Notably, Legal Thrillers stands out with the highest average ratings, indicating positive reception among readers._

## 3\. Top Authors with Multiple Bestsellers

```sql
SELECT
    book_author AS Author,
    COUNT(DISTINCT book_title) AS "Bestseller Count"
FROM
    dbo.Book_Ranks
WHERE
    book_author NOT LIKE 'Unknown'
GROUP BY
    book_author
HAVING
    COUNT(DISTINCT book_title) > 1
ORDER BY
    "Bestseller Count" DESC;
```

**Explanation:**

_This query highlights authors who have multiple bestsellers in the dataset. It is evident that Sarah J. Maas has the highest number of books in Amazon's Top 100 list. This can be attributed to her series "__A Court of Thorns and Roses" that holds the ranks of #23, #25, #47, #72, #88._

## 5\. Top Books with Highest Average Review Ratings

```sql
SELECT TOP 10
    rb.book_title,
    AVG(cr.review_rating) AS avg_review_rating
FROM
    dbo.Book_Ranks rb
JOIN
    dbo.Customer_Reviews cr ON rb.book_title = cr.book_title
GROUP BY
    rb.book_title
ORDER BY
    avg_review_rating DESC;
```
